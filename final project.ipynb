```python
#dataset source = https://nces.ed.gov/OnlineCodebook/Session/Codebook/a944d8ec-6453-44ef-944b-ab7a6dcc7173

#import and read dataset
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
file = 'Dataset-file-SSOCS/SSOCS_07_08_SSOCS08_v1_0.csv.csv'
df = pd.read_csv(file)
df.head()
```

```python
#making a pie charts for these variables - copied into google sheets (wouldnt have made sense to put into csv)
#I made one for the top 9 variables yes/no
#9 was because in a bar graph of the ranked #nos there was a sizeable gap between variable 9 and 10
```

```python
#SET STYLE FOR REST OF NOTEBOOK
matplotlib.style.use('ggplot') #set stylesheet (find options here: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (15, 12) #standardise figure size
#matplotlib.rcParams['font.family'] =  'Franklin Gothic Medium'
#matplotlib.rcParams['font.weight'] =  'bold' #<-consider adding this line (uncommented) if using a thin font
```

```python
col = 'c0116' #specifying the column name
col_name = 'Students Must Pass Through Metal Detector' #name for column (in this case the variable name)
freq = df[col].value_counts().rename_axis(col_name).reset_index(name='Frequency') #naming the columns using variable above
freq = freq.sort_values(by = col_name)

#replace 1 and 2 with yes and no
codebook = [['1', 'Yes'], #the new names i want them to have
            ['2', 'No']]

for lbl in codebook: #uses For loop to replace names, col_name is used so copy-paste would be easier
    freq[col_name] = freq[col_name].replace(int(lbl[0]), (lbl[1]))
    
freq.groupby([col_name]).sum().plot(kind='pie', y='Frequency', 
            title = col_name, figsize=(7,7), autopct='%1.1f%%')
```

```python
columns = list(df) #list columns in dataframe
table = [] #start with empty table cause its going into a function 
for c in columns:
    if (df[c].dtype) == 'int64': #when columns were listed all the yes/nos said int64
        nf = df[df[c] == 2] #specify that we want the nos (from freq chart above)
        no_count = len(nf) #find number of nos listed
        row = [c, no_count] #have the table list the column name and number of nos
        table.append(row) #add the rows listed in variables above to the table


table = pd.DataFrame(table, columns = ['variables', 'no_count']) #make table into a dataframe   
table #print table


```

```python
nnf = table.nlargest(n=31, columns=['no_count']) #print 20 largest values; n can be altered to see more rows
nnf
```

```python
nnf.to_csv('no_counts.csv')
```
